```python
import plotly.graph_objects as go
from bs4 import BeautifulSoup
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import re
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
import nltk
nltk.download('stopwords')

stemmer = PorterStemmer()

def clean_text(text):
    text = BeautifulSoup(text, "html.parser").get_text()
    text = re.sub(r'[^a-zA-Z]', ' ', text)
    text = text.lower()
    tokens = word_tokenize(text)
    stop_words = set(stopwords.words('english'))
    tokens = [word for word in tokens if word not in stop_words]
    stemmed_tokens = [stemmer.stem(word) for word in tokens]
    cleaned_text = ' '.join(stemmed_tokens)
    return cleaned_text

# Apply preprocessing to the 'description' column
raw_data['clean_description'] = raw_data['description'].apply(clean_text)

# Join all cleaned descriptions into a single string
all_descriptions = ' '.join(raw_data['clean_description'])

# Generate word frequencies
wordcloud = WordCloud(width=800, height=400, background_color='white').generate(all_descriptions)

# Create the word cloud plot
fig = go.Figure(go.Image(z=wordcloud.to_array(), hoverinfo='skip'))

# Update layout
fig.update_layout(title='Word Cloud for Descriptions')

# Show plot
fig.show()
```
